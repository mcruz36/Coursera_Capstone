```python
def getVenues(categoryID):
    CLIENT_ID = 'DMD5LLTF0QCSTADMB4HUXSXBYUUOI3XD5CXPAV5ONKJIYWBA' 
    CLIENT_SECRET = '4JMQVQF2FJIBOEHHS1RJ5DOJUFRRX243MTHA0KWSREH0KVSN' 
    VERSION = '20191201'

    CHI_LAT = 41.881832
    CHI_LONG = -87.623177

    RADIUS = 45000 # Rough radius (227 sq mi) of Chicago

    queryList = []
    offset = 0
    limit = 100
    
    venueList = []
    
    while True:
        url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&categoryId={}&limit=100&sortByDistance=1&offset={}'.format(
            CLIENT_ID,
            CLIENT_SECRET,
            VERSION,
            CHI_LAT,
            CHI_LONG,
            RADIUS,
            categoryID,
            offset
            )

        results = requests.get(url).json()
        for item in results['response']['groups'][0]['items']:
            venueList.append(item)
        
        print("Current offset: " + str(offset) + " Current Len of items: " + str(len(results['response']['groups'][0]['items'])))
        offset += limit
        
        if len(results['response']['groups'][0]['items']) != limit:
            print("Total Results: " + str(results['response']['totalResults']))
            break
        
    return venueList
```
