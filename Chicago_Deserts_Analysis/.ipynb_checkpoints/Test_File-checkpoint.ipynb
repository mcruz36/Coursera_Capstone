## Visualizing the Map

```python

from geopy.geocoders import Nominatim 
chicagoAddress = '520 S State St, Chicago, IL'


geoLocator = Nominatim(user_agent="foursquare_agent")
homeAddress = geoLocator.geocode(chicagoAddress)

loc = [homeAddress.latitude, homeAddress.longitude]
print(loc)

homeMap = folium.Map(location = loc, zoom_start = 12)

folium.GeoJson(gj, name='gj').add_to(homeMap)

# Test average 

for j in range(0,len(gj[:])):
    x = gj[j]['geometry']['coordinates'][0][0]

    cord1Sum = 0
    cord2Sum = 0
    for i in range(0,len(x)):
        cord1Sum += x[i][0]
        cord2Sum += x[i][1]

    cord1Avg = cord1Sum/len(x)
    cord2Avg = cord2Sum/len(x)

    folium.CircleMarker([cord2Avg, cord1Avg],popup = gj[j]['properties']['community'], radius = 5, color ='black' , fill = True, fill_color='red').add_to(homeMap)


homeMap
```
